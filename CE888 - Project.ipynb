```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading the JOBS dataset

```python
df_jobs = pd.read_csv('jobs.txt', sep=',')
df_jobs.head()
```

## Cleaning & Exploring - JOBS


```python
# Summary Stats
df_jobs.describe().transpose()
```

Here we have 3212 observations and 20 variables. It seems variables x3 through x6, x14, x15, and x17 are binary variables which take zero and one. Besides, variables x7 through x13, and x16 are standardised with mean of zero and standard deviation of one. So it decrease the spread of data which help us when we want to choose the model for evaluation. The mean for outcome variable is close to one, which means the number of employed people are much higher than unemployed ones, so we need to check for unbalancedness in this dataset. On the other hand the mean for treatment variable shows that only a few number of people received the treatment as it is close to zero.

```python
# Look for missing values
df_jobs.isnull().sum()
```

The sum of all missing values in this dataset is zero, so, fortunately, we do not need to deal with this issue.

```python
# Chack the type of all variables in the dataset
df_jobs.info()
```

All variables in this dataset are in type of float64, so there is no need for Encoding the categorical variables and convert them to numbers using getdummy approach.

```python
# Investigate the correlation between all the features with the outcome, treatment variable, and variable "e". 
df_jobs.corr()
```

The table of correlations, especially between features and outcome, help us to have a better understanding of the relationship between them. It is a useful information when we need to drop some irrelevant features to increase the performance and speed of the model. However, it should be considered that this is just a primitive analysis and the correlation may misslead us.

```python
# Primitive understanding of the data by generating the bargrapghs of each variable 
his=df_jobs.hist(bins=50, figsize=(20,15))
plt.savefig('Fig1.jpg')
```

The first thing that visualises what we understand from summary stats about imbalance data comes from the very taler bar for value one in the graph related to the outcome. So, in the next step, we check for it. Another piece of information that directly comes from the graphs is that variables x7 through x10, x12, x13, and x16 are right-skewed, it rings a bell for the possibility of outliers' existence.

```python
# Check for Imbalanceness in Data
df_jobs['y'].value_counts(), df_jobs['t'].value_counts()
```

```python
imb_y = len(df_jobs[df_jobs.y==1])/len(df_jobs['y'])
imb_t = len(df_jobs[df_jobs.t==1])/len(df_jobs['t'])
imb_y, imb_t
```

Here, one can easily see the amount of imbalance in the data. Around 85 percent of this sample are employed, while only nine percent of the sample received the treatment. We should consider it later when we want to split data into test-train datasets.

```python
print('The percentage of the unemployed people that did not receive training is {}%, and'.format(round(len(df_jobs[(df_jobs.y==0)
                                                                    & (df_jobs.t==0)])/len(df_jobs[df_jobs.y==0]),4)*100))

print('the percentage of the trained people that are employed is {}%'.format(round(len(df_jobs[(df_jobs.y==1)
                                                                    & (df_jobs.t==1)])/len(df_jobs[df_jobs.t==1]),4)*100))
```

The above findings are important as the number of treated persons in the sample is very small, so if we want to see the percentage of treated people with respect to all employed people, the result missleads us. However, it is better to consider two percentages: (1) what percentage of unemployed people did not receive the treatment, and (2) what percentage of the trained persons are employed. 

Based on the above sentences, we can formulate our questions as follows: "whether the treated people more likly to find a job".

```python
# Look for missing values
df_ihdp.isnull().sum()
```

```python
# Chack the type of all variables in the dataset
df_ihdp.info()
```
